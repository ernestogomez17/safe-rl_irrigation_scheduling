```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Path to the CSV file
file_path = '/home/c/cglee/egomez/irrigation_project/daily_weather_data.csv'

# Load the CSV file into a DataFrame
df = pd.read_csv(file_path)
```

```python
# Convert the 'Date' column to datetime format
df['Date'] = pd.to_datetime(df['Date'])

# Number of variables to plot (excluding 'Date')
n_vars = len(df.columns) - 1

# Create a figure with subplots
fig, axes = plt.subplots(nrows=n_vars, ncols=1, figsize=(10, 5 * n_vars), sharex=True)

# Check if there's only one variable to plot, wrap the axes in a list
if n_vars == 1:
    axes = [axes]

# Plotting each variable with Date on the x-axis
for i, column in enumerate(df.columns[1:]):
    axes[i].plot(df['Date'], df[column], linestyle='-', label=column)
    axes[i].set_title(f'{column} over Time')
    axes[i].set_xlabel('Date')
    axes[i].set_ylabel(column)
    axes[i].grid(True)

fig.tight_layout()
plt.show()
```
